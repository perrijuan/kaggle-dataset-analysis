```python
from PIL import Image
from pylab import *
from google.colab import files


# Faz o upload do arquivo
uploaded = files.upload()

# Pega o nome do arquivo enviado
nome_arquivo = list(uploaded.keys())[0]

# Abre a imagem e converte para tons de cinza
im = array(Image.open(nome_arquivo).convert('L'))

# Plota a imagem
figure()
gray()
contour(im, origin='image')
axis('equal')
axis('off')

show()
```

```python
# a taxa de cinza na imagem
figure()
hist(im.flatten(),128)
show()
```

```python
import numpy as np
im = np.array(Image.open(nome_arquivo))
print (im.shape, im.dtype)

#representamos o shape da imagem em 3dimensoes
#(altura,largura e cor) = xyz

im = np.array(Image.open(nome_arquivo).convert('L'),'f')
print(im.shape, im.dtype)
```

```python

imagem = np.array(Image.open(nome_arquivo))

#imagem esta transformada em cinza pois para reduzir a complexidade de calculode
#matriz para m x n x 3 para m x n

if len(imagem.shape) == 3:
      imagem = imagem.mean(axis=2).astype(np.uint8)



#modificamos uma regiao da imagem para criar um quadrado branco
imagem[30:100, 30:100] = 255

imagem[50, :] = 0

plt.imshow(imagem, cmap="gray")
plt.show()
```

nesse caso criamos uma modificacao na imagem atraves da manipulacao de dados dos vetores usando o numpy logo conseguimos modificar a estrutura vetorial da imagem nesse caso teriamos m x n x 3
