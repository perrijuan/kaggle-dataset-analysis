```python
import cv2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('/content/1000_F_296054899_mL331Z12tb1eHOryBlkhkX8tBFHffs3I.jpg')
img2 = cv2.imread('/content/1743535348496.jpg')
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
img1.shape
```

```python
img2.shape
```

```python
img1 = cv2.resize(img1,(800,800))
```

```python
img2 = cv2.resize(img2,(800,800))
```

```python
blended = cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta=0.2,gamma=10) #sombreamento
```

```python
plt.imshow(blended)
```
