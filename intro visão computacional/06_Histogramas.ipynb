```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ys = cv2.imread('/content/YPVz5mC.jpeg')
show_parede = cv2.cvtColor(ys, cv2.COLOR_BGR2RGB) # converte para rgb
```

```python
plt.imshow(show_parede)
```

```python
hist = cv2.calcHist([ys],channels=[2],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist)
```

```python
img = ys
```

```python
cores = ['r', 'g', 'b']
for i, col in enumerate(cores):
  histfull = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(histfull,color=col)
  plt.xlim([0,256])
  plt.ylim([0,15000])

plt.title('Histograma')
```

a imagem que foi escolida mostra altos niveis de mistrua tipicos de uma mistura e sombras
