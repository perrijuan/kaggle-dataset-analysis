```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
im = Image.open("/content/xpwallpaper.webp").convert("RGB") #converte webp para jpeg
im.save("test.jpg", "jpeg")
```

```python
img = cv2.imread('/content/test.jpg')
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #converter para rgb pois ele usa um outro padrao
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV_FULL)
plt.imshow(img)
```
